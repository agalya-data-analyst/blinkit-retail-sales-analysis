#**BLINKIT ANALYSIS**              

**Project objectives**

To conduct a comprehensive analysis of Blinkit's sales performance, customer satisfaction, and inventory distribution to identify key insights and opportunities for optimization using various KPIs and visualizations.

#####KPI goals
*  Total Revenue Assessment: To calculate and monitor the overall revenue generated from all items sold across the platform.

*  Average Transaction Evaluation: To determine the average revenue generated per sale to understand consumer spending behavior.

*  Inventory & Product Volume Tracking: To measure the total count and variety of different items sold to analyze stock movement.

*  Customer Satisfaction Mapping: To track and analyze the average customer ratings for all items sold to gauge overall service and product quality.


#####Visualisation goals
* Fat Content Impact Analysis: To analyze how the fat content of products impacts total sales and examine how other metrics like average sales, number of items, and ratings vary because of it.

* Product Performance Identification: To identify the performance of different item types in terms of total sales and their associated KPI variations.

* Outlet Segmentation Analysis: To compare total sales and key metrics across different outlets, segmented specifically by product fat content.

*  Age Evaluation: To evaluate how the age or type of an outlet establishment influences its total sales performance.

* Outlet Size Correlation: To analyze the correlation between the physical size of the outlet and the total sales it generates.

* Geographic Distribution Assessment: To assess the geographic distribution of sales across different store locations to find regional trends.

#### **Data Cleaning**

```python
#To know the unique values in  first column

print(df['Item Fat Content'].unique())
```

```python
#To replace the incorrect values in first column

df['Item Fat Content'] = df['Item Fat Content'].replace({'LF':'Low Fat',
                                                         'low fat':'Low Fat',
                                                         'reg':'Regular'})

print(df['Item Fat Content'].unique())
```

## **BUSINESS REQUIREMENTS**

##### **KPI's Requirements**

```python
#Total Sales

total_sales = df['Sales'].sum()

print(f"Total Sales :$ {total_sales:,.1f}")
```

```python
#Average Sales

avg_sales = df['Sales'].mean()

print(f"Average Sales : ${avg_sales:,.1f}")
```

```python
#Number of items sold

no_of_items_sold = df['Sales'].count()

print(f"Number of items sold : {no_of_items_sold:,.1f}")
```

```python
#Average ratings

avg_ratings = df['Rating'].mean()

print(f"Average ratings:{avg_ratings:,.1f}")
```

##### **Charts Requirements**

###### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat,labels = sales_by_fat.index,
                       autopct='%.1f%%',
                      startangle=90)

plt.title('Sales by Fat Content')
plt.axis('equal')

plt.show()
```

###### **Total Sales by Item type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))

bars=plt.bar(sales_by_type.index,sales_by_type.values)

plt.title('Total Sales by Item type')
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

###### **Fat Content by outlet for Total Sales**

```python

grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]


ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

###### **Total Sales by Outlet Establishment**

---



```python


sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()


plt.figure(figsize=(9, 5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='--')


plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment')


for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

###### **Sales by Outlet size**

```python
import matplotlib.pyplot as plt

# Grouping by Outlet Size and summing Sales
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()


plt.figure(figsize=(4, 4))
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

###### **Sales by Outlet Location**

```python

# Grouping by Outlet Location Type, summing Sales, and resetting index
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)


plt.figure(figsize=(8, 3))  # Smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)


plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()
plt.show()
```
